# Data Profiling
Generate a ydata-profiling report for the listings dimension.

```
import duckdb
import pandas as pd
from ydata_profiling import ProfileReport

conn = duckdb.connect('../data/gold/airbnb_london.duckdb')
```

```
df_listings = conn.execute("SELECT * FROM dim_listings").df()
```

```
profile = ProfileReport(df_listings, title='London Airbnb Listings Profiling Report', minimal=True)
profile.to_file('../reports/listings_profile.html')
```
