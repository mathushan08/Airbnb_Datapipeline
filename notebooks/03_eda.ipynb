# Exploratory Data Analysis (EDA)
Analyze pricing trends, property distributions, and geographic hotspots.

```
import duckdb
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

conn = duckdb.connect('../data/gold/airbnb_london.duckdb')
```

```
df = conn.execute("SELECT * FROM dim_listings l JOIN dim_geography g ON l.listing_id = g.listing_id").df()
```

```
plt.figure(figsize=(10,6))
sns.histplot(df[df['price'] < 1000]['price'], bins=50)
plt.title('Distribution of Prices in London (<$1000)')
plt.show()
```
