# Statistical Hypothesis Testing
Test differences between superhosts and regular hosts, and property types.

```
import duckdb
import pandas as pd
from scipy import stats

conn = duckdb.connect('../data/gold/airbnb_london.duckdb')
```

```
df = conn.execute("SELECT l.price, h.host_is_superhost FROM dim_listings l JOIN dim_hosts h ON l.host_id = h.host_id WHERE l.price < 2000").df()
```

```
superhost = df[df['host_is_superhost'] == True]['price'].dropna()
regular = df[df['host_is_superhost'] == False]['price'].dropna()
t_stat, p_val = stats.ttest_ind(superhost, regular, equal_var=False)
print(f'T-test p-value: {p_val}')
```
