# Predictive Modeling
Predict listing prices using XGBoost and SHAP for feature importance.

```
import duckdb
import pandas as pd
from sklearn.model_selection import train_test_split
import xgboost as xgb

conn = duckdb.connect('../data/gold/airbnb_london.duckdb')
```

```
df = conn.execute("SELECT price, accommodates, bedrooms, bathrooms, has_wifi, has_ac FROM dim_listings WHERE price < 1500").df()
```

```
df = df.dropna()
X = df.drop('price', axis=1)
y = df['price']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

model = xgb.XGBRegressor()
model.fit(X_train, y_train)
print(f'R^2 Score: {model.score(X_test, y_test)}')
```
